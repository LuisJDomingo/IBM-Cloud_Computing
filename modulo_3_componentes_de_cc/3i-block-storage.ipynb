# Almacenamiento de bloques

En este capitulo, hablaremos sobre el almacenamiento en bloque y cómo se compara con el almacenamiento de archivos en la nube. El almacenamiento en bloque divide los archivos en fragmentos (o bloques) de datos y almacena cada bloque por separado bajo una dirección única. Al igual que el almacenamiento adjunto directo y el almacenamiento de archivos, el almacenamiento en bloque también debe estar conectado a un nodo de computación antes de que pueda utilizarse para sus cargas de trabajo.

El almacenamiento en bloque, al igual que el almacenamiento de archivos, se puede montar desde dispositivos de almacenamiento remotos, lo que lo hace extremadamente resistente a fallas y mantiene los datos mucho más seguros en ellos, debido a los servicios de cifrado en tránsito y cifrado en reposo, disponibles en estos dispositivos. El almacenamiento en bloque se monta como un volumen para nodos de cómputo utilizando una red dedicada de fibras, a través del cual las señales se mueven a la velocidad de la luz.

Estas redes de fibra óptica son más caras de construir que las de Ethernet que entregan File Storage, lo que es una de las razones por las que Block Storage tiende a tener un precio más alto. Sin embargo, dado que el tráfico se mueve más rápido y con velocidad constante, son perfectos para cargas de trabajo que necesitan almacenamiento de baja latencia para funcionar de manera eficaz.

En términos de cargas de trabajo, es importante señalar que, a diferencia del almacenamiento de archivos, que puede montarse en 80 nodos de cómputo o más, Block Storage normalmente se monta en un solo nodo de cómputo a la vez. Dado que estos discos funcionan a una alta velocidad constante, son perfectos para cargas de trabajo que necesitan un almacenamiento constantemente rápido, como bases de datos y servidores de correo.

El almacenamiento en bloque no es adecuado para cargas de trabajo en las que es necesario haber algún nivel de compartición de disco entre nodos de computación. Para el almacenamiento en bloque, al igual que para el almacenamiento de archivos, es necesario tomar la capacidad de IOPS del almacenamiento en cuenta. La mayoría de los proveedores de nube le permitirán especificar características de IOPS cuando aprovisione almacenamiento, y en algunos casos, ajuste los IOPS de su almacenamiento según sea necesario, de modo que si los requisitos o el comportamiento de uso de una aplicación cambia, pueda ajustarlo en consecuencia.

Entonces, para resumir los puntos en común y las diferencias entre estos dos tipos de almacenamiento:

## Difrenceia entre Block y File storage

- El almacenamiento de bloques y archivos se obtiene de dispositivos mantenidos por el proveedor de servicios.
- Ambos son normalmente altamente disponibles y resistentes y a menudo incluyen cifrado de datos en reposo y en tránsito.
- El almacenamiento de archivos se adjunta a un nodo computacional mediante una red Ethernet, por lo que a veces se llama Almacenamiento conectado a la red o NFS.
- El almacenamiento de archivos es muy confiable, pero la velocidad de la red de conexión puede variar según la carga.
- El almacenamiento en bloque se conecta a través de una red de fibra de alta velocidad, que es muy confiable y consistente.
- El almacenamiento de archivos se puede adjuntar a varios nodos computacionales a la vez.
- El almacenamiento en bloque solo se puede conectar a un nodo a la vez.
- El almacenamiento de archivos es una buena opción cuando se requieren archivos compartidos, donde las cargas de trabajo no requieren conectividad ultrarrápida al almacenamiento, o donde el costo es un factor.
- El almacenamiento en bloque es una buena opción cuando se admite una aplicación que necesita una velocidad constante y rápida de acceso a discos, como bases de datos.

## Diferencia entre Block y File Storage

| Característica                                 | Block Storage                                               | File Storage                                                     |
|------------------------------------------------|-------------------------------------------------------------|------------------------------------------------------------------|
| Mantenimiento                                  | Dispositivos mantenidos por el proveedor de servicios         | Dispositivos mantenidos por el proveedor de servicios             |
| Disponibilidad y resistencia                   | Altamente disponible, resistente, incluye cifrado en tránsito y reposo | Altamente disponible, resistente, incluye cifrado en tránsito y reposo |
| Conexión                                       | Red de fibra de alta velocidad                               | Red Ethernet (Almacenamiento conectado a la red o NFS)            |
| Confiabilidad de la conexión                   | Muy confiable y consistente                                  | Muy confiable, pero la velocidad puede variar según la carga      |
| Conexión simultánea a nodos                    | Solo un nodo a la vez                                        | Varios nodos a la vez                                             |
| Aplicaciones recomendadas                      | Aplicaciones que necesitan acceso constante y rápido (bases de datos) | Archivos compartidos, donde la conectividad rápida no es esencial |
| Escenarios de uso                              | Alta velocidad constante para bases de datos y servidores de correo | Archivos compartidos, costo es un factor                          |


Recuerde considerar los requisitos de IOPS de la aplicación al aprovisionar almacenamiento de archivos o bloques.

>"Hola chicos, mi nombre es Amy Blea. Estoy en el equipo de ofertas aquí con IBM Cloud y hoy quiero hablarte sobre algunos tipos tradicionales de almacenamiento: almacenamiento de archivos y de bloques. Te doy una visión general de lo que son y luego darte algunos consejos sobre cómo decidir cuándo usar uno u otro.

>Entonces, comencemos con el almacenamiento en bloques. El almacenamiento en bloque es un almacenamiento donde los datos se escriben en bloques sin procesar en el almacenamiento y son accedidos por los servidores que tienes a través de una red de área de almacenamiento. Entonces, tienes todos tus servidores, pueden estar en la misma red entre sí o en una red diferente, pero todos se conectan al almacenamiento a través de esta red de área de almacenamiento.

>Ahora bien, algunas de las ventajas de utilizar el almacenamiento en bloques es que el almacenamiento en bloques proporciona la latencia más baja posible para tus aplicaciones, y también puedes usarlo para aquellas aplicaciones que requieran alto rendimiento o muchos IOPS. Otra cosa que encontrarás con el almacenamiento en bloques es que, en general, es muy redundante. La mayoría de los servicios de almacenamiento en bloque ofrecen una capacidad incorporada, donde tus datos son redundantes en todo el volumen, de modo que si hubiera un volumen que debería bajar o un disco que debería desaparecer, puedes recuperar tus datos desde otro lugar sin tener ningún tipo de impacto en tu aplicación.

>Ahora, el almacenamiento de archivos está conectado a tu servicio de manera un poco diferente a como lo está el almacenamiento en bloque. Todo está conectado en la misma red, por lo que tienes todos tus archivos o tus archivos compartidos aquí. Esos son todos los archivos y se puede acceder a todos ellos en la misma red por cualquiera de los servidores de esa red. Entonces, es almacenamiento conectado a la red.

>El almacenamiento de archivos es altamente escalable, por lo que puedes tener múltiples recursos compartidos de archivos en tu red y tener todos tus servidores conectados a él al mismo tiempo. Es accesible a múltiples tiempos de ejecución. Entonces, aquí en esta ilustración, tenemos un único archivo compartido que tiene varios servidores accediendo a todos a la vez, y también puedes tener múltiples lecturas y escrituras simultáneas en tu archivo compartido al mismo tiempo sin tener que preocuparte de que se sobrescriban tus datos.

>Entonces, ¿cómo sabes cuándo es el momento adecuado para elegir almacenamiento en bloque o almacenamiento de archivos para tu aplicación? Bueno, antes que nada, debes pensar para qué lo estás usando. Entonces, por ejemplo, digamos que tienes una configuración de VMware donde tienes varios servidores virtuales con VMware y necesitas volúmenes de arranque. Bueno, en ese caso, usarías almacenamiento en bloques. Si tienes cargas de trabajo como bases de datos transaccionales o bases de datos relacionales que requieren latencia muy baja y alto rendimiento, elegirías almacenamiento en bloque.

>En situaciones en las que tienes una combinación de datos estructurados y no estructurados, por ejemplo, un servidor de alojamiento web donde tienes tanto archivos de texto como archivos multimedia, entonces elegirías el almacenamiento de archivos. Y luego, si tienes un espacio colaborativo donde necesitas tener múltiples usuarios accediendo a todo al mismo tiempo, trabajando juntos, leyendo y escribiendo todo al mismo tiempo, elegirías el almacenamiento de archivos.

>Entonces, como mencioné al principio, el almacenamiento en bloques y en archivos son tipos de almacenamiento muy tradicionales. No son tan llamativos, brillantes y emocionantes como algunos de los nuevos servicios de almacenamiento que han entrado en el mercado, pero siguen siendo muy relevantes y útiles para los diferentes tipos de cargas de trabajo que tengas, ya sea local o en la nube.
