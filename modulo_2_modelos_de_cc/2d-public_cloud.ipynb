## Introducción

En nuestro vídeo introductorio sobre la nube, mencionamos brevemente los **cuatro modelos de implementación para la nube**.  
En este vídeo, analizaremos con más detalle el **modelo de implementación de la nube pública**.

Los modelos de implementación indican:
- Dónde reside la infraestructura,
- Quién la posee y la administra,
- Cómo los recursos y servicios de la nube se ponen a disposición de los usuarios.

## Modelos de Implementación de la Nube

Los cuatro modelos de implementación de la nube incluyen:  
1. **Nube pública**  
2. **Nube privada**  
3. **Nube comunitaria**  
4. **Nube híbrida**

### Nube Pública

En un modelo de **nube pública**, los usuarios obtienen acceso a:
- Servidores
- Almacenamiento
- Redes
- Seguridad
- Aplicaciones

Todo esto es proporcionado como **servicios entregados por proveedores de servicios en la nube a través de Internet**.  
Mediante consolas web y API, los usuarios pueden **suministrar los recursos y servicios que necesitan**.

El proveedor de la nube:
- **Posee, gestiona, aprovisiona y mantiene la infraestructura**, alquilándola a los clientes por un cargo de suscripción o una tarifa basada en el uso.

### Propiedad y Gestión de los Recursos

Los usuarios no son propietarios de:
- Los servidores en los que se ejecutan sus aplicaciones
- El almacenamiento que consumen sus datos
- La gestión de las operaciones de los servidores
- La determinación de cómo se mantienen las plataformas

**Similar a los servicios públicos** como agua, electricidad o gas, no somos propietarios de los recursos en la nube. Hacemos un acuerdo con el proveedor de servicios, usamos los recursos y pagamos por lo que usamos en un período determinado.

### Beneficios de la Nube Pública

Las nubes públicas ofrecen **importantes ahorros de costes** en términos de **coste total de propiedad (TCO)**, ya que:
- El proveedor corre con todos los gastos de capital, operación y mantenimiento de la infraestructura y las instalaciones.

Esto también facilita la escalabilidad, haciendo que sea tan sencillo como **solicitar más capacidad**.

### Limitaciones de la Nube Pública

Sin embargo, con una nube pública:
- El usuario **no tiene control sobre el entorno informático**.
- Está sujeto al rendimiento y la seguridad de la infraestructura del proveedor de la nube.

## Proveedores de Nube Pública

Actualmente, existen varios proveedores de nube pública en el mercado, tales como:
- **Amazon Web Services**
- **Microsoft Azure**
- **IBM Cloud**
- **Google Cloud Platform**
- **Alibaba Cloud**

Si bien todos los proveedores incluyen un conjunto común de servicios básicos como servidores, almacenamiento, redes, seguridad y bases de datos, también ofrecen un **amplio espectro de servicios especializados** con variadas opciones de pago.

## Características de una Nube Pública

1. Una **nube pública** es una arquitectura virtualizada de múltiples inquilinos que permite a los usuarios compartir recursos informáticos fuera de sus **firewalls**.
   
2. El conjunto de recursos del proveedor de la nube, incluida la infraestructura, las plataformas y el software, **NO está dedicado** para uso de un solo inquilino u organización.

3. Los recursos se distribuyen según sea necesario y se ofrecen a través de una variedad de **suscripciones y modelos de pago por uso**.

## Beneficios de la Nube Pública

- **Amplios recursos bajo demanda**, lo que permite que las aplicaciones respondan sin problemas a las fluctuaciones en la demanda.
- Debido a la gran cantidad de usuarios compartiendo recursos centralizados, la nube pública ofrece **economías de escala**.
- La gran cantidad de recursos de servidor y red disponibles en la nube pública garantiza **alta disponibilidad**. Si un componente físico falla, el servicio aún se ejecuta sin verse afectado.

## Preocupaciones en la Nube Pública

A pesar de sus beneficios, las nubes públicas tienen algunas preocupaciones importantes, como:
1. **Seguridad**:
   - Violaciones de datos
   - Pérdida de datos
   - Secuestro de cuentas
   - Insuficiente diligencia
   - Vulnerabilidades del sistema y de las aplicaciones

2. **Soberanía de los datos**:  
   Los datos se almacenan en diferentes ubicaciones y se accede a ellos a través de fronteras nacionales. Por ello, es crucial para las empresas **cumplir con las regulaciones de soberanía de datos** que rigen el almacenamiento, la transferencia y la seguridad de los datos.

La capacidad del proveedor para mantenerse al día con las regulaciones y su interpretación es una preocupación compartida por muchas empresas.

## Casos de Uso Comunes de la Nube Pública

- Las organizaciones optan por **aplicaciones y plataformas basadas en la nube** para que sus equipos puedan centrarse en desarrollar y probar aplicaciones, reduciendo el tiempo de comercialización.
- Empresas con necesidades fluctuantes de recursos eligen la nube pública.
- Organizaciones están utilizando **recursos de computación en la nube pública** para construir infraestructuras secundarias para recuperación de desastres, protección de datos y continuidad del negocio.
- Cada vez más organizaciones utilizan servicios de **gestión de datos y almacenamiento en la nube** para una mayor accesibilidad, distribución y respaldo de sus datos.
- Departamentos de TI están subcontratando la gestión de **plataformas y aplicaciones menos críticas** a proveedores de nube pública.
