# Bienvenido a ¿Qué es la seguridad en la nube, parte 2?

Las organizaciones están migrando su carga de trabajo a un **entorno de nube híbrida**, lo que aumenta la **superficie de ataque** de amenazas. Esta expansión puede generar nuevos desafíos de **cumplimiento** y **seguridad de los datos**. Por lo tanto, necesita un programa sólido de **ciberseguridad centrado en los datos** para proteger sus datos. 

Además, necesita **visibilidad y monitoreo centralizados** contra accesos no autorizados, exposición o robo de datos en todo el entorno de datos de su empresa. Analicemos algunas capacidades de seguridad de datos que puede implementar para proteger sus datos.

## Mejoras en la seguridad de los datos

Para proteger los datos, puede mejorar la madurez de las personas, los procesos y la tecnología:

- **Identifique sus activos de datos más críticos**: Sepa quién tiene acceso a ellos y cómo están protegidos.
- Evite la pérdida de datos detectando, previniendo y haciendo cumplir las infracciones de políticas.
- **Implemente la gobernanza de la seguridad de los datos** estableciendo procesos, métricas y un estado continuo de descubrimiento y clasificación de datos.
- **Monitoree la seguridad de la base de datos** aplicando políticas de protección de datos y cumplimiento en entornos de nube híbrida.

## Gestión de accesos y autenticación

Su próximo enfoque en un entorno de nube debería ser la **gestión de los controles de acceso** y la **autenticación**. Necesita desarrollar estrategias de **gestión de acceso e identidades (IAM)** en la nube que trabajen bajo una arquitectura de **confianza cero**. Esto protege contra riesgos y asegura que cualquier usuario que acceda a un recurso sea autenticado constantemente.

### Al construir la nube de IBM, considere:

1. **Modernícese a su propio ritmo**: Adapte los requisitos a su negocio.
2. **Conserve sus inversiones existentes y aplicaciones locales** mientras incorpora herramientas de **Arquitectura IAM** en la nube.
3. **Implemente confianza cero**: Controle el acceso centralizado, preserve la confidencialidad del cliente, reduzca amenazas internas y proteja los recursos remotos.
4. **Escalabilidad de IAM**: Asegure que su IAM en la nube sea capaz de escalar para millones de usuarios o transacciones sin modificar la infraestructura.

## Seguridad de la red en la nube

Otro elemento clave de la seguridad en la nube es la **seguridad de la red en la nube**, que implica medidas, tecnologías, políticas, controles y procesos para proteger las redes de nube pública, privada e híbrida.

### Beneficios de la seguridad de red en la nube:

- **Monitoreo y gestión centralizados**.
- Gestión de políticas granular y actualizable.
- **Detección en tiempo real** y **prevención contra intrusiones**, DDoS y otros ataques.
- Configuración y gestión automatizadas para evitar errores.
- **Servicios de cifrado** para proteger los datos en reposo y en tránsito.
- Gestión centralizada de identidad y acceso.

## Mejores prácticas para la seguridad en la nube

Siga estos pasos críticos en **tres fases** para mejorar la seguridad en la nube:

### Fase 1: Identificar el estado de uso de la nube y sus riesgos

- Identifique cómo se accede a sus datos.
- Detecte el uso desconocido de la nube.
- Verifique las configuraciones de sus servicios en la nube.
- Monitoree signos de uso malicioso de datos en la nube.

### Fase 2: Proteger su sistema en la nube

- Asigne políticas de protección.
- **Cifre datos sensibles**.
- Formule políticas para el intercambio de datos.
- Restringa el intercambio de datos a dispositivos desconocidos.
- Implemente soluciones de mitigación y protección de bots.
- Use soluciones adecuadas de antimalware.

### Fase 3: Responder a ataques

- Agregue pasos adicionales de autenticación y verificación para escenarios de acceso de alto riesgo.
- Agregue nuevas políticas para nuevos servicios en la nube.

## Principios de seguridad en la nube del NIST

El **Instituto Nacional de Estándares y Tecnología (NIST)** ha establecido un marco de ciberseguridad que incluye cinco pilares clave: **Identificar**, **Proteger**, **Detectar**, **Responder** y **Recuperar**.

## Gestión de la postura de seguridad en la nube (CSPM)

Las soluciones de **CSPM** abordan errores comunes, como configuraciones incorrectas, ayudando en la implementación de componentes clave de seguridad en la nube, como **gestión de identidad y acceso (IAM)**, gestión de cumplimiento normativo, monitoreo de tráfico, respuesta a amenazas y gestión de activos digitales.

---

El aumento masivo en la adopción de la nube está impulsando a los ciberdelincuentes a atacar entornos en la nube. Aunque las organizaciones están tomando medidas de seguridad, siguen siendo vulnerables a los ciberataques. Las principales tendencias en seguridad en la nube incluyen **estrategias multinube**, **modelos de confianza cero**, **entornos híbridos y multinube**, **DevSecOps**, seguridad para la fuerza laboral remota e inteligencia artificial y **aprendizaje automático** para la detección de amenazas.
