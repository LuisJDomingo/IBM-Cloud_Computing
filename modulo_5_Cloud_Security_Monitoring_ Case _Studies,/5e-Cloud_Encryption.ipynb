Dadas las preocupaciones sobre la **seguridad y privacidad de los datos**, especialmente en entornos de nube pública, el **cifrado** juega un papel clave y a menudo se hace referencia a él como la **última línea de defensa** en un modelo de seguridad en capas. Esta protección no solo cifra los datos, sino que también proporciona un **sólido control de acceso**, **gestión de claves** y **gestión de certificados**.

En este vídeo, veremos más de cerca el **cifrado en la nube**. El cifrado se define como **codificar datos de manera que sean ilegibles**. Un sistema de cifrado consta de dos partes:

1. **El algoritmo de cifrado**: Define las reglas mediante las cuales los datos se transformarán para que se vuelvan ilegibles.
2. **La clave de descifrado**: Define cómo se volverán a transformar los datos cifrados en datos legibles.

El cifrado garantiza que solo los **usuarios autorizados** tengan acceso a los datos confidenciales y que, cuando se accede a ellos o se interceptan sin autorización, los datos sean **ilegibles** y **sin sentido**.

### Servicios de cifrado en la nube

Los **proveedores de la nube** ofrecen varios servicios de cifrado en la nube. Esto puede variar desde un **cifrado limitado de datos** identificados como confidenciales hasta un **cifrado de extremo a extremo** de todos los datos cargados en la nube. Los datos se cifran al recibirlos y las **claves de cifrado** se pasan a los clientes para que puedan descifrar los datos cuando sea necesario. Es fundamental que las **claves se gestionen de forma segura**, ya que si se pierden, no podrás leer tus datos.

---

### Tres estados de protección de datos

Los datos necesitan protección en tres estados:

1. **En reposo**: Protege los datos mientras están almacenados físicamente en una base de datos o en la capa de almacenamiento.
   - Existen múltiples opciones para cifrar datos en reposo, como el cifrado para almacenamiento de archivos y bloques, servicios de cifrado integrado, almacenamiento de objetos y cifrado de bases de datos.

2. **En tránsito**: Protege los datos mientras se transmiten de una ubicación a otra.
   - Incluye cifrar los datos antes de la transmisión, autenticar puntos finales y descifrar/verificar los datos a su llegada.
   - Protocolos como **SSL** (Capa de Sockets Seguros) y **TLS** (Seguridad de la Capa de Transporte) son comúnmente utilizados.

3. **En uso**: Protege los datos cuando se utilizan en la memoria para cálculos, permitiendo realizar cálculos sobre texto cifrado sin necesidad de descifrar los datos.

---

### Tipos de cifrado

1. **Cifrado del lado del servidor**:
   - Ocurre después de que el almacenamiento en la nube recibe los datos, pero antes de que se escriban en el disco y se almacenen.
   - Puedes optar por:
     - **Claves de cifrado proporcionadas por el cliente**: Crear y gestionar tus propias claves.
     - **Claves de cifrado administradas por el cliente**: Utilizar los servicios de gestión de claves del proveedor de almacenamiento en la nube.

2. **Cifrado del lado del cliente**:
   - Ocurre antes de que los datos se envíen al almacenamiento en la nube.
   - Los usuarios pueden utilizar **claves de cifrado y algoritmos** que no son visibles para el proveedor de la nube, haciendo prácticamente imposible que el proveedor descifre los datos alojados.

### Estrategia de cifrado en entornos de múltiples nubes

Dado que la mayoría de las empresas operan en **entornos de múltiples nubes**, es necesario implementar una **estrategia singular de protección de datos** en implementaciones locales, híbridas y de múltiples nubes. Algunos proveedores de la nube ofrecen **servicios de cifrado de datos en múltiples nubes** con una variedad de características, como:

- **Gestión de acceso a datos**,
- **Gestión integrada de claves**,
- **Cifrado sofisticado**, ofreciendo escalabilidad y flexibilidad para proteger las cargas de trabajo más sensibles en toda la empresa, independientemente de dónde residan los datos.

Usando una **consola de cifrado de datos de múltiples nubes**, puedes definir y gestionar políticas de acceso, crear, rotar y gestionar claves de cifrado, y **agregar registros de acceso**.

---

### Gestión de claves

El cifrado no elimina el riesgo de seguridad de los datos, sino que **traslada la seguridad a las claves de cifrado**. Estas claves deben ser gestionadas y protegidas contra amenazas para mantener los datos seguros.

Los **servicios de gestión de claves** ofrecidos por algunos proveedores de la nube ayudan a realizar el ciclo de vida de la gestión de claves de cifrado que se utilizan en servicios en la nube o en aplicaciones creadas por el cliente. Estos servicios permiten a los clientes:

- Cifrar datos confidenciales en reposo,
- Crear y gestionar fácilmente todo el ciclo de vida de las claves criptográficas que se utilizan para cifrar datos.

Dado que las **llaves permanecen en posesión del cliente**, los datos están protegidos tanto de los proveedores de servicios en la nube como de otros usuarios.

---

### Mejores prácticas para la gestión de claves de cifrado

Algunas **mejores prácticas** incluyen:

- **Almacenar claves de cifrado** por separado de los datos cifrados.
- Realizar copias de seguridad clave fuera del sitio y auditarlas periódicamente.
- Actualizar las claves periódicamente.
- Implementar autenticación multifactor para las claves maestras y de recuperación.
