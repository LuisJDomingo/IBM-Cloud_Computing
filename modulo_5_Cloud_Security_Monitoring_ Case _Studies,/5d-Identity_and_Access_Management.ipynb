Según el **Informe de seguridad en la nube** elaborado por *Cybersecurity Insiders*, la principal preocupación de seguridad en la nube para los profesionales de la ciberseguridad es la **pérdida y fuga de datos**.

El acceso no autorizado mediante el uso indebido de las credenciales de los empleados y los **controles de acceso inadecuados** son las vulnerabilidades percibidas más grandes para la seguridad en la nube, seguido de **interfaces y API inseguras**.

En este video, veremos cómo la **gestión de identidad y acceso**, también conocida como control de acceso, funciona como primera línea de defensa, permitiéndote autenticar y autorizar a los usuarios y proporcionar acceso específico a recursos, servicios y aplicaciones en la nube.

---

Una **estrategia de seguridad integral** debe abarcar las necesidades de una amplia audiencia, incluyendo:

- Usuarios de las organizaciones,
- Usuarios de Internet y redes sociales,
- Organizaciones,
- Proveedores de socios comerciales externos.

### Existen tres tipos principales de usuarios:

1. **Usuarios administrativos**: Incluyen administradores, operadores y gerentes de plataformas en la nube, roles que normalmente crean, actualizan y eliminan instancias de aplicaciones y servicios, y también necesitan información sobre las actividades de los miembros de su equipo.

2. **Usuarios desarrolladores**: Incluyen a los desarrolladores de aplicaciones en la nube, desarrolladores de plataformas y editores de aplicaciones. Estos usuarios están autorizados a leer información confidencial, crear, actualizar y eliminar aplicaciones.

3. **Usuarios de aplicaciones**: Son los usuarios de las aplicaciones alojadas en la nube.

---

### Componentes clave de la gestión de identidades y accesos:

1. **Autenticación**: Permite que las aplicaciones desplegadas en la nube autentiquen a los usuarios a nivel de aplicación, basadas en una variedad de proveedores de identidad (directorio en la nube, identidad social como Google, LinkedIn, Facebook, Twitter, o identidad alojada en la empresa).

2. **Autenticación multifactor (MFA)**: Utilizada para combatir el robo de identidad, agregando una capa adicional de autenticación para los usuarios de la aplicación (contraseñas o PINs de un solo uso, certificados, tokens, autenticación basada en riesgos).

3. **Servicios de Cloud Directory**: Gestionan de forma segura los perfiles de usuario, sus credenciales y políticas de contraseñas dentro de un entorno de nube.

4. **Informes**: Proporcionan una visión centrada en el usuario del acceso a los recursos o una visión centrada en los recursos del acceso de los usuarios.

5. **Auditoría y cumplimiento**: Servicios críticos para validar los controles implementados contra la política de seguridad de la organización, el cumplimiento de la industria y las políticas de riesgo.

---

### Gestión del acceso de usuarios y servicios:

Permite aprovisionar y desaprovisionar perfiles de usuarios, clientes, socios y proveedores con mínima interacción humana. Esto agiliza el control de acceso basado en roles, organización y políticas de acceso.

Las cuentas de usuario de **administradores y desarrolladores** dan acceso a información confidencial. Para mitigar los riesgos de hackeo, necesitas el máximo control sobre todo el ciclo de vida de estos usuarios, incluyendo:

- Aprovisionar usuarios especificando roles en los recursos,
- Políticas de contraseñas (caracteres especiales, longitud mínima),
- Autenticación multifactor (contraseñas de un solo uso basadas en el tiempo),
- Cancelación inmediata del acceso cuando los usuarios dejan o cambian de rol.

---

### Servicios de gestión de identidad y acceso de proveedores de la nube:

Incluyen la capacidad de:

- Crear **grupos de acceso**,
- Agregar usuarios a grupos,
- Gestionar el acceso de usuarios existentes.

Un **grupo de acceso** es un conjunto de usuarios e ID de servicio creado para asignar el mismo acceso a todas las entidades dentro del grupo mediante una o más políticas de acceso.

Las **políticas de acceso** definen cómo los usuarios, ID de servicio y grupos de acceso reciben permisos para acceder a los recursos de la cuenta.
